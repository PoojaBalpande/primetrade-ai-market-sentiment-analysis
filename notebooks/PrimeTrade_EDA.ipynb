# 5. Data Cleaning & Preprocessing

The datasets are free from missing values and duplicate records. Therefore, preprocessing focuses on improving data consistency and preparing the datasets for analysis.

The preprocessing steps include:

- Converting timestamps into datetime format
- Standardizing date formats
- Verifying data types
- Creating additional features for analysis

```python
# Create copies of original datasets

historical_clean = historical_df.copy()
sentiment_clean = sentiment_df.copy()
```

```python
# Convert date column

sentiment_clean["date"] = pd.to_datetime(sentiment_clean["date"])

print("Fear & Greed date converted successfully.")
```

```python
historical_clean["Timestamp"] = pd.to_datetime(
    historical_clean["Timestamp"],
    unit="ms"
)
```

```python
historical_clean["Timestamp IST"] = pd.to_datetime(
    historical_clean["Timestamp IST"],
    dayfirst=True
)
```

```python
historical_clean["Timestamp"].head()
```

```python
historical_clean[["Timestamp", "Timestamp IST"]].head()
```

```python
sentiment_clean["date"].head()
```

## Feature Engineering

To prepare the datasets for analysis, we create additional features that simplify merging and enable time-based insights.

The new features include:

- Trading Date
- Trading Hour
- Trading Day
- Market Sentiment Date

```python
# Extract only the date from Timestamp IST

historical_clean["Trade Date"] = historical_clean["Timestamp IST"].dt.date

historical_clean[["Timestamp IST", "Trade Date"]].head()
```

```python
historical_clean["Trade Hour"] = historical_clean["Timestamp IST"].dt.hour

historical_clean["Trade Hour"].head()
```

```python
historical_clean["Day of Week"] = historical_clean["Timestamp IST"].dt.day_name()

historical_clean["Day of Week"].head()
```

```python
sentiment_clean["Trade Date"] = sentiment_clean["date"].dt.date

sentiment_clean[["date", "Trade Date"]].head()
```

```python
print(historical_clean[["Trade Date", "Trade Hour", "Day of Week"]].dtypes)

print()

print(sentiment_clean[["Trade Date"]].dtypes)
```

# 6. Merge Datasets

The historical trading data is merged with the Bitcoin Fear & Greed Index using the common **Trade Date**.

This allows every trade to be associated with the market sentiment on the day it occurred.

```python
merged_df = historical_clean.merge(
    sentiment_clean[
        ["Trade Date", "classification", "value"]
    ],
    on="Trade Date",
    how="left"
)

merged_df.head()
```

```python
merged_df.rename(
    columns={
        "classification": "Market Sentiment",
        "value": "Fear & Greed Value"
    },
    inplace=True
)

merged_df.head()
```

```python
print("=" * 70)
print("Merged Dataset")
print("=" * 70)

print("Rows    :", merged_df.shape[0])
print("Columns :", merged_df.shape[1])

print()

print("Missing Market Sentiment:",
      merged_df["Market Sentiment"].isnull().sum())
```

```python
merged_df.to_csv(
    "../data/processed/merged_dataset.csv",
    index=False
)

print("Merged dataset saved successfully!")
```

# 7. Exploratory Data Analysis (EDA)

## 7.1 Market Sentiment Distribution

Before analyzing trader performance, it is important to understand the distribution of Bitcoin market sentiment over the trading period.

This analysis answers:

- Which market sentiment occurs most frequently?
- Is the dataset balanced across different sentiment categories?
- What market conditions dominate the historical data?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count market sentiment
sentiment_counts = (
    merged_df["Market Sentiment"]
    .value_counts()
    .sort_values(ascending=False)
)

print(sentiment_counts)

plt.figure(figsize=(8,5))

ax = sns.countplot(
    data=merged_df,
    x="Market Sentiment",
    order=sentiment_counts.index
)

plt.title("Distribution of Market Sentiment", fontsize=15, fontweight="bold")
plt.xlabel("Market Sentiment")
plt.ylabel("Number of Trades")
plt.xticks(rotation=15)

# Add labels
for container in ax.containers:
    ax.bar_label(container)

plt.tight_layout()
plt.show()
```

### Observations

- The chart shows the frequency of each Bitcoin market sentiment category.
- This provides context for understanding trader behaviour under different market conditions.
- Any imbalance in sentiment distribution should be considered while interpreting subsequent analyses.

## 7.2 Distribution of Closed Profit and Loss (PnL)

Closed PnL represents the realized profit or loss from completed trades.

Analyzing its distribution helps identify:

- Overall profitability
- Presence of outliers
- Spread of trader performance

```python
plt.figure(figsize=(10,5))

sns.histplot(
    merged_df["Closed PnL"],
    bins=60,
    kde=True
)

plt.title("Distribution of Closed Profit & Loss")
plt.xlabel("Closed PnL")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,2))

sns.boxplot(
    x=merged_df["Closed PnL"]
)

plt.title("Boxplot of Closed PnL")

plt.tight_layout()
plt.show()
```

### Observations

- The distribution indicates whether trader profits are evenly distributed or dominated by a few large trades.
- The boxplot highlights potential outliers that may significantly influence the average profit.

## 7.3 Average Trader Profit by Market Sentiment

This analysis evaluates whether trader profitability changes under different Bitcoin market sentiment conditions.

Average Closed PnL is calculated for each sentiment category.

```python
avg_pnl = (
    merged_df
    .groupby("Market Sentiment")["Closed PnL"]
    .mean()
    .sort_values(ascending=False)
)

display(avg_pnl)
```

```python
plt.figure(figsize=(8,5))

ax = sns.barplot(
    x=avg_pnl.index,
    y=avg_pnl.values
)

plt.title("Average Closed PnL by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Average Closed PnL")

plt.xticks(rotation=15)

for container in ax.containers:
    ax.bar_label(container, fmt="%.2f")

plt.tight_layout()
plt.show()
```

### Business Insights

- This visualization directly measures trader performance under different market sentiment conditions.
- Higher average PnL indicates more profitable trading environments.
- These findings help determine whether traders perform better during Fear or Greed periods.

```python
merged_df["Winning Trade"] = merged_df["Closed PnL"] > 0

win_rate = (
    merged_df
    .groupby("Market Sentiment")["Winning Trade"]
    .mean()*100
)

display(win_rate)
```

```python
plt.figure(figsize=(8,5))

ax = sns.barplot(
    x=win_rate.index,
    y=win_rate.values
)

plt.title("Winning Trade Percentage by Market Sentiment")
plt.ylabel("Win Rate (%)")
plt.xlabel("Market Sentiment")

for container in ax.containers:
    ax.bar_label(container, fmt="%.1f%%")

plt.tight_layout()
plt.show()
```

### Business Insights

- Traders achieved the highest average realized profit during **Extreme Greed** market conditions (67.89).
- Surprisingly, **Fear** also resulted in relatively high average profits (54.29), indicating that experienced traders may capitalize on market downturns.
- Trading performance was comparatively lower during **Greed** and **Neutral** market conditions.
- Overall, trader profitability appears to vary with market sentiment, suggesting that sentiment is an important factor influencing trading outcomes.

```python
plt.figure(figsize=(9,5))

sns.countplot(
    data=merged_df,
    x="Market Sentiment",
    hue="Side"
)

plt.title("Buy vs Sell Across Market Sentiment")

plt.tight_layout()
plt.show()
```

```python
top_coins = merged_df["Coin"].value_counts().head(10)

plt.figure(figsize=(10,6))

ax = sns.barplot(
    x=top_coins.values,
    y=top_coins.index
)

plt.title("Top 10 Most Traded Coins")

for container in ax.containers:
    ax.bar_label(container)

plt.tight_layout()
plt.show()
```

```python
numeric_cols = [
    "Execution Price",
    "Size Tokens",
    "Size USD",
    "Closed PnL",
    "Fee",
    "Fear & Greed Value"
]

corr = merged_df[numeric_cols].corr()

plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Matrix")

plt.tight_layout()
plt.show()
```

### Business Insights

- Transaction fees increase significantly with trade size, as expected.
- Closed PnL exhibits only a weak correlation with trade size and execution price, indicating that profitability depends on multiple factors beyond position size.
- The Fear & Greed Index does not show a strong linear relationship with numerical trading variables, suggesting that market sentiment may influence trading behavior in more complex, non-linear ways.

```python
plt.figure(figsize=(8,5))

volume = merged_df.groupby("Market Sentiment")["Size USD"].sum().sort_values(ascending=False)

ax = sns.barplot(x=volume.index, y=volume.values)

plt.title("Total Trading Volume by Market Sentiment")
plt.ylabel("Total Volume (USD)")
plt.xticks(rotation=15)

for container in ax.containers:
    ax.bar_label(container, fmt="%.0f")

plt.tight_layout()
plt.show()
```

```python
coin_profit = (
    merged_df.groupby("Coin")["Closed PnL"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,6))

ax = sns.barplot(
    x=coin_profit.values,
    y=coin_profit.index
)

plt.title("Top 10 Coins by Total Closed PnL")

for container in ax.containers:
    ax.bar_label(container, fmt="%.0f")

plt.tight_layout()
plt.show()
```

```python
daily_trades = merged_df.groupby("Trade Date").size()

plt.figure(figsize=(12,5))

daily_trades.plot()

plt.title("Daily Trading Activity")
plt.xlabel("Date")
plt.ylabel("Number of Trades")

plt.tight_layout()
plt.show()
```
